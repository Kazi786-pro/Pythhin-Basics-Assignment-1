```python
# 15 >>>> Write a Python program to calculate the factorial of a number provided
# by the user using a while loop.

n = int(input("Enter any number: "))

f = 1
while n >= 1:
    f *= n
    n -= 1

# print output
print("Factorial is", f)
```
